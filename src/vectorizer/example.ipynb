```python
from datasets import load_dataset

train_dataset = load_dataset("swan07/authorship-verification", split="train")
test_dataset = load_dataset("swan07/authorship-verification", split="test")

# train_dataset = load_dataset("swan07/authorship-verification", split="train[:5000]")
# test_dataset = load_dataset("swan07/authorship-verification", split="test[:1000]")
```

```python
from specter import SpecterVectorizer

spectorVectorizer = SpecterVectorizer()
spectorVectorizer.load()

# spectorVectorizer.embed(train_dataset)
spectorVectorizer.embed(test_dataset)
```

```python
from bert import BertVectorizer

bertVectorizer = BertVectorizer()
bertVectorizer.load()

# bertVectorizer.embed(train_dataset)
bertVectorizer.embed(test_dataset)
```

```python
from tf_idf import TfIdfVectorizer

tfVectorizer = TfIdfVectorizer()

tfVectorizer.train(train_dataset)
tfVectorizer.save("tfidf_vectorizer_model.pkl")

# tfVectorizer.load("tfidf_vectorizer_model.pkl")
# tfVectorizer.embed(train_dataset)
tfVectorizer.embed(test_dataset)
```

```python
from ngram import NgramCountVectorizer

ngramVectorizer = NgramCountVectorizer()

ngramVectorizer.train(train_dataset)
ngramVectorizer.save("ngram_vectorizer_model.pkl")

# ngramVectorizer.load("ngram_vectorizer_model.pkl")
# ngramVectorizer.embed(train_dataset)
ngramVectorizer.embed(test_dataset)
```
